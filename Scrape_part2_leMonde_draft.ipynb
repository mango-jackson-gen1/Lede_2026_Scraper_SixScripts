```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
response = requests.get("https://www.lemonde.fr/en/#")
doc = BeautifulSoup(response.text, 'html.parser')
print(response.status_code)   # 200 means OK
```

```python
stories = doc.find_all(class_= "article")
```

```python
len(stories)
```

find(class or a tag name / h1, h2)

select = css selector 

select_one = multiples options 

```python
#find,
#selc
```

```python
rows = []
for story in stories:
    row = {}

    row ['title'] = story.find(class_ ="article__title").text.strip()
    
    try:
        row['link'] = story.find('a')['href']
    except:
        row['link'] = None

    try:
        row['premium'] = story.find(class_ = 'sr-only').text
    except:
        row['premium'] = None
    
    
    try: 
        row['subhead'] = story.find(class_ = 'article__desc').text.strip()
    except:
        row['subhead'] = None

    try:
        row['byline'] = story.find(class_ = 'article__byline').text.strip()
    except:
        row['byline'] = None
    
    img = story.find("img")
    try:
       row['image_url'] = story.find('img')['src']
    except:
       row['image'] = None

    try:
        row['article_type'] = story.find(class_ = 'article__type').text
    except:
        row['article_type'] = None





    rows.append(row)


len(rows)

```

```python
import pandas as pd
df = pd.DataFrame(rows)
```
