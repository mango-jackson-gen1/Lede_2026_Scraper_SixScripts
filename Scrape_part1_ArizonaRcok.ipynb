```python
import requests
from bs4 import BeautifulSoup
url = "https://rocktumbler.com/blog/rock-and-mineral-clubs/"
headers = {"User-Agent": "Mozilla/5.0"}  # some sites block requests without this

response = requests.get(url, headers=headers)
doc = BeautifulSoup(response.text, "html.parser")
print(response.status_code)   # 200 means OK
```

```python
table2 = doc.find_all("table")[5]
```

```python
entries = table2.find_all('tr')
```

```python
table2.find('tr')
print(table2.find('tr').prettify()) 
```

Add in the names1

```python
clubs = []

for entry in entries:
    cells = entry.find_all('td')      # [name_cell, city_cell]

    name_cell = cells[0]
    city_cell = cells[1]              # the <td width="40%"> with the city

    name_print = name_cell.text.strip()
    a_tag = name_cell.find('a')
    link1 = a_tag.get('href', None) if a_tag else None
    city = city_cell.text.strip()     # "Apache Junction"

    clubs.append({
        "name": name_print,
        "link": link1,
        "city": city,
    })

    print("-----")
    print(name_print)
    print(link1)
    print(city)

```

WHen ready to be saved, it should be a LIST of Dictionaries
