```python
import curl_cffi

url = 'https://travel-id-documents.az.gov/authorized-third-party-driver-license-locations'

r = curl_cffi.get(url, impersonate='chrome') 
#same as requests.get() but a different handshake for https connection 

print(r.status_code)            # 200 = worked
```

```python
#<table>
    #<tbody>
        #<tr>
            #<td>
                #<a href> for the link
                #<a href>. text
            #<td>  

        #tr
        #tr     
```

```python
# 2. Parse the HTML
doc = BeautifulSoup(r.text, 'lxml')

# 3. Loop the data rows - go to the lowest level just before the for loop 
table = doc.find('table')
entries = table.find('tbody').find_all('tr')
```

```python
<td>Check Changers of Arizona LLC<br>
			&nbsp;</td>
```

```python
rows = []
# skip the first <tr> because it's the header, not data
for entry in entries:
    row = {}
    
    try:
        row['name'] = entry.find('a').text.strip()
    except:
        row['name'] = entry.find('td').text.strip()   
    # no <a> → grab the first <td>'s text
    #find('td') gives back the first matching one. no index needed 


    try:
        row['link'] = entry.find('a')['href']
    except:
        row['link'] = None

    # row['name']   = cells[0].text.strip()
    # row['link'] = cells[1].text.strip()
  

    # link = cells[0].find('a')  #already at the <a> link level
    # #no need to wrtie as find('a')['href']

    # try:
    #     row['link'] = link['href'] 
    # except:
    #     row['link'] = None


    # # link = cells[0].find('a')
    # # row['link'] = link['href'] if link else None
    
    rows.append(row)

len(rows)
```
